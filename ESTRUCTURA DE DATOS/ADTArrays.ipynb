```python
"""
ADT Array.
Estatico
"""
class Array():
  def __init__(self , tamano):
    self.__tamano = tamano
    self.__data = [0 for x in range(self.__tamano)]          # Generador

  def get_item( self , index ):
    return self.__data[index]

  def to_string( self ):
    print(self.__data)
```

```python
ejemplo = Array(5)
ejemplo.to_string()
ejemplo.get_item(3)
```
